```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Функція для зменшення фігури в 1.5 рази
def reduce_figure(vertices):
    centroid = np.mean(vertices, axis=0)  # Знаходимо центроїд
    reduced_vertices = (vertices - centroid) * (2/3) + centroid  # Зменшуємо в 1.5 рази
    return reduced_vertices

# Задаємо вершини довільного опуклого 8-кутника
vertices = np.array([[1, 3], [3, 1], [6, 1], [8, 3], [7, 5], [5, 7], [3, 6], [2, 5]])

# Додаємо першу вершину в кінець масиву для замикання фігури
vertices = np.vstack([vertices, vertices[0]])

# Зменшуємо фігуру в 1.5 рази
reduced_vertices = reduce_figure(vertices)

# Виводимо вершини зменшеного 8-кутника та вершини початкового 
print("Зменшені вершини 8-кутника:")
for vertex in reduced_vertices:
    print(vertex)

print('Початкові вершини:')
print(vertices)
# Відображаємо початковий 8-кутник
plt.figure(figsize=(8, 8))
plt.plot(vertices[:, 0], vertices[:, 1], 'bo-', label='Початковий 8-кутник')
plt.fill(vertices[:, 0], vertices[:, 1], alpha=0.5)
for i, vertex in enumerate(vertices[:-1]):
    plt.text(vertex[0], vertex[1], f'V{i+1}', ha='right')

# Відображаємо зменшений 8-кутник
plt.plot(reduced_vertices[:, 0], reduced_vertices[:, 1], 'ro-', label='Зменшений 8-кутник')
plt.fill(reduced_vertices[:, 0], reduced_vertices[:, 1], alpha=0.3)
for i, vertex in enumerate(reduced_vertices):
    plt.text(vertex[0], vertex[1], f'V{i+1}', ha='right')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('8-кутник')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()


###### Пункт про симетрію

plt.figure(figsize=(10, 10))
plt.plot(vertices[:, 0], vertices[:, 1], 'bo-', label='Початковий 8-кутник')
plt.fill(vertices[:, 0], vertices[:, 1], alpha=0.5)
for i, vertex in enumerate(vertices[:-1]):
    plt.text(vertex[0], vertex[1], f'V{i+1}', ha='right')
plt.plot(symmetric_reduced_vertices[:, 0], symmetric_reduced_vertices[:, 1], 'yo-', label='Симетричний зменшений 8-кутник')
plt.fill(symmetric_reduced_vertices[:, 0], symmetric_reduced_vertices[:, 1], alpha=0.4)
for i, vertex in enumerate(symmetric_reduced_vertices):
    plt.text(vertex[0], vertex[1], f'V{i+1}', ha='right')
plt.plot(reduced_vertices[:, 0], reduced_vertices[:, 1], 'ro-', label='Зменшений 8-кутник')
plt.fill(reduced_vertices[:, 0], reduced_vertices[:, 1], alpha=0.3)
for i, vertex in enumerate(reduced_vertices):
    plt.text(vertex[0], vertex[1], f'V{i+1}', ha='right')

plt.title('Восьмикутники')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.xlim(-9, 9)
plt.ylim(-9, 9)
plt.legend(loc='lower right')
plt.axhline(0, color='black',linewidth=0.7)
plt.axvline(0, color='black',linewidth=0.7)
plt.show()
```
